```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('../data/processed/sales.db')

df = pd.read_sql('SELECT * FROM sales', conn)

df.head()
```

Faturamento por mês

```python
df.groupby('Month')['Revenue'].sum().sort_index()
```

Dá para observar um aumento significativo no faturamento nos meses de novembro e dezembro, indicando vendas muito concentradas em certas épocas, possivelmente associada a datas comerciais como Black Friday e fim de ano.
A empresa pode concentrar investimentos em marketing e estoque nesses períodos para aumentar o lucro.

Categoria que mais vende

```python
df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
```

A categoria Technology representa a maior parcela do faturamento, indicando uma demanda maior dos produtos dessa categoria. 
Estratégias comerciais podem priorizar produtos de tecnologia, com campanhas ou um investimento maior em novos produtos.

Top 10 produtos

```python
df.groupby('Product Name')['Revenue'].sum().sort_values(ascending=False).head(10)
```

Tem produtos com faturamento muito maior que outros. Como existe essa concentração de receita em poucos produtos, acaba indicando uma dependência de itens específicos.
